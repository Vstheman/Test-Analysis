```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_main = pd.read_csv("../input/prosple-dat-analysis/Trial_ds_data.csv")
df=df_main
print(df.shape)
df.head(5)
```

### Cleaning the data

```python
print(df.isnull().sum().sum(),"Null Values")
df.isnull().sum() # Null values
```

```python
df.dtypes
```

```python
# Replacing numerical NAN values with the mean of that column
column_means = df.mean()
df = df.fillna(column_means)

# Replacing NAN values in categorical columns with the most occuring element
cat_cols = [x for x in df if df[x].dtypes == "object"]
for col in cat_cols:
    mode_ = df[col].mode()[0]
    df = df.fillna(mode_)
    
# The data had a random date as an entry in multiple cells so those rows are being dropped along with the whole date column 
df = df.iloc[: , 1:]
df = df.replace("21/04/2022", np.nan)
df.dropna(axis=0, inplace=True)
```

```python
# All unique entries in every column
cats = [x for x in df if df[x].dtypes == "object"]
for x in cats:
    print(x, "\n",df[x].unique())
    print("-----------------------------------------------\n")
```

```python
df.corr()
```

```python
# Subtracting the mean of the survey scores of people with 1 and 2 years of experience  
print(df.loc[df["How long have you been with your current employer?"].isin(["1 year"])].mean() - df.loc[df["How long have you been with your current employer?"].isin(["2 years"])].mean())
```

```python
# Selecting the top 11 sectors as per the number of entries
df["Company Sector"].value_counts()
df_sects = df.loc[df["Company Sector"].isin(["Accounting & Advisory", "Government & Public Service","Engineering Consulting","Technology","Retail, Sales & Consumer Goods","Law","Banking & Financial Services","Construction & Property Services","Energy & Utilities","Mining, Oil & Gas"])]
df_sects["Company Sector"].value_counts()
```

```python
arr_sect = df_sects["Company Sector"].unique()
for x in arr_sect:
    print(x,"\n",df.loc[df["Company Sector"].isin([x])].mean(),"----"*20,"\n")
```

## Insights
